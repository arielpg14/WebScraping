```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # Para graficación
from sklearn.cluster import KMeans # Para el modelamiento
```

```python
# Ubicacion del archivo de datos
url = '/content/covidSouthAmerica.csv' # external url?
```

```python
# Se carga el archvio csv con pandas DataFrame
data = pd.read_csv(url) # Recupera los datos en formato csv a tipo dataframe (Estructura de datos bidimensional)
data.head() # Visualiza los 5 primeros datos del dataframe
```

```python
# Visualizamos el DataFrame
data
```

```python
# Remove unnecesary cols
# https://www.educative.io/answers/how-to-delete-a-column-in-pandas
# https://sparkbyexamples.com/pandas/pandas-drop-multiple-columns-by-index
# data.drop('a', inplace=True, axis=1)
data = data.drop(data.columns[[0]],axis = 1)
data
```

```python
df = data
# Reemplazar "no data" por 0 en las columnas relevantes
columnas_relevantes = ['Country/Territory', 'Cases', 'Active cases', 'Deaths', 'Recoveries']
df[columnas_relevantes] = df[columnas_relevantes].replace('NaN', 0)
# Mostrar el DataFrame resultante
print(df)
```

```python
df = data
# Especifica la columna en la que deseas reemplazar NaN por 0
columna_con_nan = 'Recoveries'
# Reemplazar NaN por 0 en la columna específica
df[columna_con_nan] = df[columna_con_nan].fillna(0)
# Mostrar el DataFrame resultante
print(df)
```

```python
# Guardamos las datos extraidos como archivo CSV
# se puede aplcair otras opciones para guardar (Base de datos)
df.to_csv('covidSouthAmerica1.csv')
```
